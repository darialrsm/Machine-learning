15. Розв'язати систему лінійних рівнянь Ax=b, де А = матриця коефіцієнтів, b - вектор вільних членів

```python
a = np.array(([1, 3, -4], [0, 2, -3], [-2, 2, 5]))
b = np.array([5, 3, 4])

x = np.linalg.solve(a, b)
print(x)

inv_a = np.linalg.inv(a)
x1 = inv_a.dot(b)
print(x1)
```

22. Сортування рядків або стовпців матриці
