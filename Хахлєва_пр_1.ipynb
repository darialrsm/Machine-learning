1. Створіть одновимірний масив (вектор) з добутками парних чисел на їхні попередні
непарні числа, починаючи з 2, та виведіть його значення. (перші 10 парних чисел)

```python
# Перші 10 парних чисел
even_numbers = np.arange(2, 21, 2)

# Попередні непарні числа
odd_numbers = even_numbers - 1

# Добутки парних та непарних чисел
result = even_numbers * odd_numbers

print("Парні числа: ", even_numbers)
print("Попередні непарні числа: ", odd_numbers)
print("Результат (вектор):", result)
```

4. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в
діапазоні від 0 до 1 та обчисліть суму елементів, які менше середнього значення.


```python
# Створюємо матрицю заданими параметрами, значення округлені до сотих

matrix = np.round(np.random.rand(5, 5), 2)
print("Початкова матриця")
print(matrix)

# Середнє значення
mean = np.round(np.mean(matrix), 2)
print("Середнє значення: ", mean)

# Значення, що менші за середнє
less_than_mean = matrix[matrix < mean]
print("Значення, що менші за середнє: ", less_than_mean)

# Обчислення суми
sum = np.sum(less_than_mean)
print("Сума значень, що менші за середнє: ", sum)
```

12. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в
діапазоні від 1 до 5 та знайдіть добуток елементів діагоналей обох матриць (окремо).

```python
# Створюємо матриці
m1 = np.random.randint(1, 6, (3, 3))
m2 = np.random.randint(1, 6, (3, 3))

# Елементи діагоналей
diag1 = np.diag(m1)
diag2 = np.diag(m2)

# Добуток елементів діагоналі
prod_diag1 = np.prod(diag1)
prod_diag2 = np.prod(diag2)

print("Матриця 1:")
print(m1)
print("Матриця 2:")
print(m2)

print("Діагональні елементи матриці 1:", diag1)
print("Діагональні елементи матриці 2:", diag2)

print("Добуток елементів діагоналі матриці 1:", prod_diag1)
print("Добуток елементів діагоналі матриці 2:", prod_diag2)
```

13. Створіть дві матриці розміром 3x3, заповніть їх випадковими дійсними числами в
діапазоні від -1 до 1 та знайдіть суму їх елементів

```python
# Створюємо матриці
m1 = np.round(np.random.uniform(-1, 1, (3, 3)), 2)
m2 = np.round(np.random.uniform(-1, 1, (3, 3)), 2)

# Знаходимо суму елементів кожної матриці
sum1 = np.sum(m1)
sum2 = np.sum(m2)

# Загальна сума
total = sum1+sum2

print("Матриця 1:")
print(m1)
print("Матриця 2:")
print(m2)

print("Сума елементів матриці 1:", sum1)
print("Сума елементів матриці 2:", sum2)

print("Загальна сума елементів двох матриць", total)
```

15. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить кількість нульових
елементів в кожному рядку матриці

```python
# Створюємо матрицю
matrix = np.round(np.random.uniform(-0.5, 0.5, (3, 3,)), 2)

zero_counts = np.sum(matrix == 0, axis=1).reshape(3, 1)

print("Матриця:")
print(matrix)

print("Вектор-стовпчик з кількістю нульових елементів у кожному рядку:")
print(zero_counts)
```
